```python
import transformers

model = transformers.AutoModelForQuestionAnswering.from_pretrained('Intel/bert-base-uncased-sparse-85-unstructured-pruneofa')
```

```python
from numpy import array
from numpy import count_nonzero
import numpy as np
```

```python
for param in model.parameters():
    np_array = param.data.numpy()
    #np_array = np.nan_to_num(np_array,0)
    sparsity = 1.0 - ( count_nonzero(np_array) / float(np_array.size) )
    print(np_array.shape,sparsity)
```
